```python
## Converting Datetime to standard format

import re
import pandas as pd

def extract_first_date(text):
    if pd.isna(text):
        return None
    
    # Remove ordinal suffixes (st, nd, rd, th)
    text = re.sub(r'(\d+)(st|nd|rd|th)', r'\1', text)
    
    # Extract first date pattern
    match = re.search(r'\d{1,2}[-/ ]?[A-Za-z]{3,9}[-/ ]?\d{2,4}|\d{1,2}/\d{1,2}/\d{4}', text)
    
    if match:
        return match.group(0)
    else:
        return None

# Apply extraction
events_df["Clean_Date"] = events_df["Date"].apply(extract_first_date)

# Convert to datetime safely
events_df["Clean_Date"] = pd.to_datetime(
    events_df["Clean_Date"],
    dayfirst=True,
    errors="coerce"
)

events_df[["Date", "Clean_Date"]].head()
```

```python
# Extract Hour
master_df["Hour"] = master_df.index.hour

# Day of Week (0 = Monday, 6 = Sunday)
master_df["Day_of_Week"] = master_df.index.dayofweek

# Month
master_df["Month"] = master_df.index.month

# Day of Year
master_df["Day_of_Year"] = master_df.index.dayofyear

# Is Weekend (Saturday=5, Sunday=6)
master_df["Is_Weekend"] = master_df["Day_of_Week"].isin([5, 6]).astype(int)

# Is Peak Hour (6 PM to 10 PM → 18 to 21)
master_df["Is_Peak_Hour"] = master_df["Hour"].between(18, 22).astype(int)
```

```python
import numpy as np

# Hour Cyclical Encoding
master_df["Hour_sin"] = np.sin(2 * np.pi * master_df["Hour"] / 24)
master_df["Hour_cos"] = np.cos(2 * np.pi * master_df["Hour"] / 24)

master_df[["Hour", "Hour_sin", "Hour_cos"]].head()
```

```python
# 1-day lag (same time yesterday)
master_df["Load_lag_1day"] = master_df["LOAD"].shift(96)

# 7-day lag (same time last week)
master_df["Load_lag_7day"] = master_df["LOAD"].shift(672)

# Short-term memory
master_df["Load_lag_1"] = master_df["LOAD"].shift(1)
master_df["Load_lag_4"] = master_df["LOAD"].shift(4)

master_df[["LOAD", "Load_lag_1day", "Load_lag_7day"]].head(110)
```

```python
# 24-hour rolling mean (96 intervals)
master_df["Rolling_mean_24h"] = master_df["LOAD"].rolling(window=96).mean()

# 4-hour rolling mean (16 intervals)
master_df["Rolling_mean_4h"] = master_df["LOAD"].rolling(window=16).mean()

# 24-hour rolling standard deviation (volatility)
master_df["Rolling_std_24h"] = master_df["LOAD"].rolling(window=96).std()

master_df[["Rolling_mean_24h", "Rolling_std_24h"]].head(110)
```

```python
def calculate_penalty(y_true, y_pred):
    penalty = 0
    
    for actual, forecast in zip(y_true, y_pred):
        if actual > forecast:
            penalty += (actual - forecast) * 4
        else:
            penalty += (forecast - actual) * 2
            
    return penalty

val_penalty = calculate_penalty(y_val.values, y_val_pred)

print("Validation Penalty:", val_penalty)
```

```python
best_penalty = float("inf")
best_shift = 0

for shift in range(-10, 21):  # try shifts from -10 to +20
    shifted_pred = y_val_pred + shift
    penalty = calculate_penalty(y_val.values, shifted_pred)
    
    if penalty < best_penalty:
        best_penalty = penalty
        best_shift = shift

print("Best Shift:", best_shift)
print("Best Penalty:", best_penalty)
```

```python
def calculate_stage2_penalty(y_true, y_pred, datetime_index):
    
    penalty = 0
    peak_penalty = 0
    offpeak_penalty = 0
    
    for actual, forecast, dt in zip(y_true, y_pred, datetime_index):
        
        hour = dt.hour
        is_peak = (hour >= 18) and (hour < 22)
        
        if actual > forecast:  # Under-forecast
            
            if is_peak:
                cost = (actual - forecast) * 6
            else:
                cost = (actual - forecast) * 4
                
        else:  # Over-forecast
            cost = (forecast - actual) * 2
        
        penalty += cost
        
        if is_peak:
            peak_penalty += cost
        else:
            offpeak_penalty += cost
            
    return penalty, peak_penalty, offpeak_penalty
```
